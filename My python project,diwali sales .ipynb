```python
import numpy as np

import pandas as pd
import matplotlib as mpl
import seaborn as s
```

```python
#import csv file 
df = pd.read_csv('diwali sales data.csv',encoding = 'Unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head(10)
```

```python
#drop blank columns 
```

```python
df.drop (['Status','unnamed1'],axis=1, inplace=True )
```

```python
df.info()
```

```python
#remove null values
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# To change datatype 
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
#To Rename any Columns
```

```python
df.rename(columns={'Marital_status':'Shadi'})
```

```python
df.describe()
```

```python
# use describe for specific columns
```

```python
df[['Age','Amount']].describe()
```

```python
#EXPLORATORY DATA ANALYSIS
```

```python
df.columns
```

```python
ax = s.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
sale_gn=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
s.barplot(x='Gender',y='Amount',data=sale_gn)
```

```python
# AGE
```

```python
df.columns
```

```python
ax =s.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age= df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
s.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
# STATE
```

```python
df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=True)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=True)
s.set(rc={'figure.figsize':(19,5)})
s.barplot(x='State',y='Orders',data=sales_state)
```

```python
# STATE AND AMOUNTS
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
s.set(rc={'figure.figsize':(19,5)})
s.barplot(x='State',y='Amount',data=sales_state)
```
